```python
import pandas as pd
import numpy as np
import os
# from nltk.tokenize import word_tokenize
```

```python
df=pd.read_csv('./GTEx Portal.csv')
df
```

```python
path = '/project/GutIntelligenceLab/ss4yd/gtex_data/actual_useful/'
svs_path_list = [os.path.join(path, x) for x in os.listdir(path)]
pathdf=pd.DataFrame(svs_path_list, columns=['paths'])
pathdf['patient_id']=pathdf.paths.apply(lambda x: x.split('/')[-1].split('.')[0])
pathdf
```

```python
pathdf['paths'][0]
```

```python
findf=pathdf.merge(df, left_on='patient_id', right_on='Tissue Sample ID')
findf
```

```python
findf.columns
```

```python
findf.columns = ['paths', 'patient_id', 'tid', 'tissue', 'sid',
       'sex', 'age', 'hardy_scale', 'cat',
       'notes']
```

```python
fincdf=findf.drop(columns=['tid', 'sid', 'hardy_scale', 'sex', 'age'])
```

```python
fincdf['maj_tissue']=fincdf['tissue'].apply(lambda x: x.split('-')[0])
```

```python
fincdf.maj_tissue.value_counts()
```

```python
fincdf.to_csv('./starting.csv', index=False)
```
